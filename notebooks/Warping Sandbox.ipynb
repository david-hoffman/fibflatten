```python
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.ndimage as ndi
import skimage.data as data
```

```python
img = data.camera()
```

```python
plt.matshow(img)
```

```python
import numpy as np
```

```python
coords = np.indices(img.shape)
```

```python
center = np.array((256,256))[:, None, None]
coords2 = np.tensordot(np.array([[np.cos(np.pi/4), np.sin(np.pi/4)], [-np.sin(np.pi/4), np.cos(np.pi/4)]]), coords - center, axes=1) + center
```

```python
def plot_coords(coords):
    plt.matshow(coords[0])
    plt.contour(coords[0], colors="k")
    plt.matshow(coords[1])
    plt.contour(coords[1], colors="k")
```

```python
plot_coords(coords)
plot_coords(coords2)
```

```python
img2 = ndi.map_coordinates(img, coords2)
```

```python
plt.matshow(img2)
```
