```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from skimage.feature import canny
from skimage.external import tifffile as tif
```

```python
data = tif.imread("testdata.tif")
```

```python
data.shape
```

```python
plt.matshow(data[300])
```

```python
c = canny(data[300], sigma=1, high_threshold=10, low_threshold=0)
plt.matshow(c)
```

```python
import scipy.ndimage as ndi
```

```python
plt.matshow(ndi.morphology.binary_erosion(ndi.morphology.binary_dilation(c, iterations=4), iterations=5))
```
